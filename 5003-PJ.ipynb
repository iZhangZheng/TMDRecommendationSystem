# 数据集的一些事情
* 第一批所有的数据都来自kaggle，链接如下：
    * [The Movies Dataset](https://www.kaggle.com/rounakbanik/the-movies-dataset) 目前打算主要是用这个数据库做，后面的三个作为补充。
    * [TMDB 5000 Movie Dataset](https://www.kaggle.com/tmdb/tmdb-movie-metadata/data)
    * [MovieLens 20M Dataset](https://www.kaggle.com/grouplens/movielens-20m-dataset)
    * [350 000+ movies from themoviedb.org](https://www.kaggle.com/stephanerappeneau/350-000-movies-from-themoviedborg)
* 数据库中很多电影数据来自"The Movie Database(TMDb)"，点击[传送门](https://www.themoviedb.org)跳转。

```scala
#header=True，表示的是第一行用作表头。
#escape='"'：忽略"内部的分隔符。
#multiLine=True：对部分内容支持多行，但是在这里似乎用不到。
df = spark.read.csv('moviedata/TMD/movies_metadata.csv',header=True,inferSchema=True,multiLine=True,escape='"')
df.show()
```

```scala
#show里面的括号是显示的项目的数量，如果需要很多项目，就在里面输入大一些的数字。
df.where("vote_count>50").select("title","vote_average").orderBy("vote_average",ascending=0).show(50)
```

```scala
df2 = spark.read.csv('moviedata/TMD/ratings.csv',header=True,inferSchema=True,multiLine=True,escape='"')
df2.show()
```

```scala
df2.groupBy("userId").count().orderBy("userId",ascending=0).show()
```
